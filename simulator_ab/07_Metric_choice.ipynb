# 2. Как выбрать метрику

Метрика должна отображать то, что для вас важно. Что для вас важно вам нужно решить самим.

Если хотите увеличить выручку с текущих клиентов, хорошей метрикой будет LTV или средняя выручка с клиента.

Если увеличить количество платящих клиентов, можно использовать конверсию в покупку.

Если хотите и конверсию и выручку, можно создавать комбинированные метрики. Рассмотрим пример в конце семинара.

#### Кейс 1. Метрика, которую сложно придумать

Нужно оценить изменение на стриминговой платформе с музыкой, типа Яндекс Музыки или Spotify. Основной источник дохода — это годовые подписки. Хотим чтобы пользователи покупали подписки снова и снова. Измерять выручку напрямую долго, так как текущая подписка у большинства пользователей закончится не скоро, а измерить эффект изменения нужно сейчас. В таких ситуациях в качестве метрики эксперимента используют «счастье пользователя». Счастье пользователя — сущность, которую сложно формализовать, но это приходится делать, чтобы оценивать успешность экспериментов. Созданием такой метрики, как правило, занимается отдельный аналитик или команда аналитиков, которые исследуют пользовательское поведение и пытаются придумать формулу, которая показывает, насколько пользователь доволен сервисом. Как это происходит? Они находят факторы, влияющие на «счастье пользователя». Например, чем больше часов в день пользователь слушает музыку с нами, тем больше ему нравится сервис. Чем разнообразнее репертуар исполнителей, которых слушает пользователь, тем больше ему нравится сервис. Как часто возвращается к прослушанным композициям и как часто досрочно переключается на следующую. Продолжительность прослушивания и разнообразие репертуара мы можем посчитать и по ним вычислить некий общий показатель счастья. Как именно это сделать ответа нет, это творческая исследовательская задача.

#### Кейс 2. Неочевидная ошибка

У нас есть поле, на котором выращивают капусту. Капусту продают по определённой цене за килограмм. Хотим больше заработать и ставим задачу вывести сорт капусты, который будет давать больший урожай. Метрика - суммарный вес капусты, собранной с поля. Вырастили столько же кочанов капусты сколько и прежде, каждый кочан оказался в среднем в 2 раза больше. То есть наша метрика увеличилась в 2 раза, можно сказать, что эксперимент успешный. Однако, продать получилось только треть урожая. Как выяснилось позже люди неохотно покупают большие кочаны, так как их неудобно нести домой и сложно съесть весь кочан, пока он не испортится. Если бы мы знали это заранее, то можно было бы заменить метрику на количество кочанов при условии что они весят столько же.

# 3. Как выбрать критерий

Выбираем статистическую гипотезу.

Статистическая гипотеза - это некоторое предположение о свойствах и характеристиках исследуемых генеральных совокупностей.

Примеры:
- выборки независимы
- данные из одного распределения
- данные из нормального распределения
- средние равны
- квантили равны
- дисперсии равны

Для каждой гипотезы есть свои критерии. Вообще критериев очень много. Можно придумывать свои.

В итоге вам нужно принять решение: внедрять или не внедрять. Тут важны вероятности ошибок.

Чтобы выбрать критерий оцените вероятности ошибок критериев и выберете тот, чьи вероятности ошибок лучше.

### Гипотеза о равенстве средних 

Мы рассматривали 3 критерия:
- тест Стьюдента
- тест Манна-Уитни
- бутстреп

Как выбрать критерий с точки зрения вероятностей ошибок обсудили ранее.

При прочих равных рекомендуем использовать тест Стьюдента, так как тест Манна-Уитни в общем случае проверяет другую гипотезу, а бутстреп вычислительно затратный.

Бывают подводные камни. Например, если у вас метрика отношения (ratio metric), то применение классических способов в лоб будет некорректным. Про работу с метриками отношения в курсе будет отдельная лекция.

Если гипотеза не равенство средних, то, как правило, можно использовать бутстреп. Либо искать специальные тесты, подходящие для вашей задачи.

# 4. Бинарные метрики

Бинарные метрики хороши тем, что у них известен вид распределения и нет выбросов.

Гипотеза о равенстве долей бинарной метрики - это частный случай проверки гипотезы о равенстве средних.

Всё что работает для проверки гипотезы о равенстве средних работает и для гипотезы о равенстве долей. Можно применять тест Стьюдента. Можно применять способы повышения чувствительности, которые будем изучать в следующих уроках.

# Конструирование метрики

Проводим эксперимент в продукте, которые только запустился и активно развивается. Важно не просто максимизировать выручку, но и получить большую базу платящих клиентов. Экспериментируем с коммуникацией с пользователем в первую неделю после установки приложения. Ожидаем что в среднем каждый клиент, совершивший хоть какую-то покупку в течение первой недели, принесёт в будущем 400 рублей. В качестве измерений метрики будем использовать сумму выручки с клиента плюс ценность факта покупки.

В текущем варианте 30% клиентов совершают покупку в среднем на 1500 рублей, в экспериментальном варианте 40% клиентов совершают покупку в среднем на 1000 рублей.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

```python
def get_values(mu, std, prob, sample_size):
    """Генерирем данные о стоимостях покупок."""
    values = np.random.normal(mu, std, sample_size).astype(int)
    mask = np.random.binomial(1, prob, sample_size)
    return values * mask

def calc_metric(values, extra):
    """Добавляем к купившим клиентам константу extra."""
    return values + (values > 0) * extra

def use_ttest(values_a, values_b):
    """Применяем t-test."""
    delta = values_b.mean() - values_a.mean()
    _, pvalue = ttest_ind(values_a, values_b)
    print(f'delta = {delta:0.0f}, pvalue = {pvalue:0.3f}')
```

```python
mu_a = 1500
mu_b = 1000
prob_a = 0.3
prob_b = 0.4
std = 100
sample_size = 5000
extra = 400

values_a = get_values(mu_a, std, prob_a, sample_size)
values_b = get_values(mu_b, std, prob_b, sample_size)
use_ttest(calc_metric(values_a, extra), calc_metric(values_b, extra))
```

Получили статистически незначимые отличия.

Как зависит результат оценки эксперимента в зависимости от дополнительной ценности купивших клиентов (extra)?

```python
for extra in range(0, 2000, 400):
    print(f'extra = {extra: <5}', end=', ')
    use_ttest(calc_metric(values_a, extra), calc_metric(values_b, extra))
```

Выбор метрики влияет на результат эксперимента.
